```python
import pandas as pd
from sklearn.decomposition import NMF
import numpy as np
```

```python
import codecs
file_name = 'drinkSAJ.csv'
with codecs.open(file_name, "r", "Shift-JIS", "ignore") as file:
    df = pd.read_table(file, delimiter=",")
# df = pd.read_csv(file_name)
df
```

```python
df[["Mu15","M1620","M2130","M3140","Mo41","Fu15","F1620","F2130","F3140","Fo41"]]
```

```python
data = df[["Mu15","M1620","M2130","M3140","Mo41","Fu15","F1620","F2130","F3140","Fo41"]]
n_components = 6

model = NMF(n_components=n_components, init='random', random_state=0)
W = model.fit_transform(data)
H = model.components_


# print("R-P×Q^Tは")
# print(model.reconstruction_err_ )
```

```python
print("Wは")
pd.DataFrame(W,index=df["Drink"])
```

```python
print("H^Tは")
pd.DataFrame(H,columns=df.columns[1:])
```

```python
print("W×H^Tは")
pd.DataFrame(np.dot(W,H))
```
